Get median-income and median home value

```
medianincome = soup.findAll(attrs={'class':'median-income'})
medianhomeval = soup.findAll(attrs={'class':'med-home-val'})
fips = soup.findAll(attrs={'class':'fips'})

print(medianincome[0].string)


```

**Convert** text to number

```
medianincome_num = [int(m.string) for m in medianincome] 
medianhomeval_num = [int(m.string) for m in medianhomeval] 
fips_num = [int(mn.string) for mn in fips] 

print(medianincome_num[0])
```

```
import numpy as np
import matplotlib.pyplot as plt

plt.scatter(fips_num,medianhomeval_num, label="median home val")
plt.scatter(fips_num,medianincome_num, label="median income")
```
