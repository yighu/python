```python
my_float = float(my_int)
print('value: {}, type: {}'.format(my_float, type(my_float)))
```

## [`decimal.Decimal`](https://docs.python.org/3/library/decimal.html)

```python
from decimal import Decimal
```

```python
from_float = Decimal(0.1)
from_str = Decimal('0.1')
print('from float: {}\nfrom string: {}'.format(from_float, from_str))
```

```python
my_decimal = Decimal('0.1')
sum_of_decimals = my_decimal + my_decimal + my_decimal
print(sum_of_decimals == Decimal('0.3'))
```

## Operator precedence in calculations
Mathematical operator precedence applies. Use brackets if you want to change the execution order:

```python
print(1 + 2**2 * 3 / 6) 

# 1 + 4 * 3 / 6 == 1 + 12 / 6 == 1 + 2
print((1 + 2**2) * 3 / 6)
```

multiple Assignment

```python
a,b,c=1,2,3
print("numbers {},{},{}".format(a,b,c))
print("a={}".format(a))
```

```python
a, b = 100, 200

print(a)
# 100

print(b)
# 200
```

```python
a, b, c = 0.1, 100, 'string'

print(a)
# 0.1

print(b)
# 100

print(c)
# string
```

```python
a = b = 100

print(a)
# 100

print(b)
# 100
```

```python
a = b = [0, 1, 2]

print(a is b)
# True

a[0] = 100
print(a)
# [100, 1, 2]

print(b)
# [100, 1, 2]
```

```python
a, *b = 100, 200, 300

print('a={}'.format(a))
print(type(a))
# 100
# <class 'int'>

print(b)
print(type(b))
# [200, 300]
# <class 'list'>
```

```python
*a, b = 100, 200, 300,50

print(a)
print(type(a))
# [100, 200]
# <class 'list'>

print(b)
print(type(b))
# 300
# <class 'int'>
```
